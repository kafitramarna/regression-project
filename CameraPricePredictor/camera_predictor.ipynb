# Import Dataset

```python
dataset = pd.read_csv('/kaggle/input/1000-cameras-dataset/camera_dataset.csv')
dataset.head()
```

```python
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X.shape)
```

# Taking care of missing data

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(X[:, :])
X[:, :] = imputer.transform(X[:, :])
print(X)
```

# Split Data Into Training Set and Test Set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

# Train Model With Multiple Linear Regression

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
```

```python
y_pred = lin_reg.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

# Train Model With Support Vector Regression

```python
from sklearn.svm import SVR
svr_reg = SVR(kernel = 'rbf')
svr_reg.fit(X, y)
```

```python
y_pred = svr_reg.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

# Train Model With Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor
dec_tree_reg = DecisionTreeRegressor(random_state = 0)
dec_tree_reg.fit(X, y)
```

```python
dec_tree_pred = dec_tree_reg.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((dec_tree_pred.reshape(len(dec_tree_pred),1), y_test.reshape(len(y_test),1)),1))
```

# Train Model With Random Forest Regression 

```python
from sklearn.ensemble import RandomForestRegressor
rand_for_reg = RandomForestRegressor(n_estimators = len(dataset), random_state = 0)
rand_for_reg.fit(X, y)
```

```python
rand_for_pred = rand_for_reg.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((rand_for_pred.reshape(len(rand_for_pred),1), y_test.reshape(len(y_test),1)),1))
```
